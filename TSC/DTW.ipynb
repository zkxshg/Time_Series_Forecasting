```python
import numpy as np
import pandas as pd
from dtw import *
from numpy.linalg import norm
from numpy import array
```

```python
help(dtw)
```

```python
T = pd.read_excel('T.xlsx')
T1 = pd.read_excel('T1.xlsx')

ser = np.array(T.iloc[:,1])
ser1 = np.array(T1.iloc[:,1])
```

```python
alignment = dtw(ser, ser1, keep_internals=True)
alignment.plot(type="threeway")
```

```python
blockNum = int(ser.size / (ser1.size))
blockNum
```

```python
for block in range(0, blockNum):
    serBlock = ser[block * ser1.size : (block + 1) * ser1.size]
    alignmentB = dtw(serBlock, ser1, keep_internals=True)
    alignmentB.plot(type="threeway")
```
